```python
class fresh(Dataset):
  def __init__(self, df):
    self.rws = df['review']
    self.labels = df['sentiment']
  def __len__(self):
    return len(self.labels)
  def __getitem__(self, index):
    return self.rws[index], self.labels[index]
```

```python
def padding_fn(batch):
  reviews,labels = zip(*batch)
  rgv = pad_sequence(reviews,batch_first = True,padding_value = 0)
  rgv = torch.tensor(rgv, dtype = torch.long)
  labels = torch.tensor(labels,dtype = torch.long)
  return rgv,labels



```
